```python
import pandas as pd
from transformers import AutoModel, AutoTokenizer, pipeline
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data/_Post_Twitter_all_Hashtags_with_matches_clean_removed_stopwords_no_lem.csv")
```

```python
df = df.applymap(str)
```

```python
list_of_tweets = list(df["Tweet_clean"])
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
# Defining a function to call for the whole dataframe
def BERT_sentiment(inpText):
  return(classifier(inpText)[0]['label'])
```

```python
df['Sentiment']=df['Tweet_clean'].apply(BERT_sentiment)
df
```

```python
df.to_csv("sentiment_100k.csv")
```

```python
# Visualizing the overall sentiment distribution
import matplotlib.pyplot as plt
fig, subPlot = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
fig.suptitle("Sentiment analysis")
 
# Grouping the data
grouped_data=df.groupby('Sentiment').size()
 
# Creating the charts
grouped_data.plot(kind='bar', ax=subPlot[0], color=['crimson', 'lightblue'])
grouped_data.plot(kind='pie', ax=subPlot[1], colors=['crimson', 'lightblue'])
```
